```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
ratings_csv = pd.read_csv('csv/Ratings.csv')
books_csv = pd.read_csv('csv/Books.csv')
users_csv = pd.read_csv('csv/Users.csv')
```

```python
df1 = ratings_csv.merge(books_csv,on='ISBN')
```

```python
df1
```

```python
num_rating = df1.groupby(['Book-Title']).count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating
```

```python
avg_rating =df1.groupby(['Book-Title']).agg({'Book-Rating': 'mean'})
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating
```

```python
popular_df = num_rating.merge(avg_rating,on='Book-Title')
```

```python
popular_df = popular_df.query('num_rating>250').sort_values(by='avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books_csv,on='Book-Title').drop_duplicates('Book-Title')
```

```python
popular_df.head()
```

```python
popular_df.drop(['ISBN','Image-URL-S','Image-URL-L'],axis=1,inplace=True)
```

```python
df2 = df1.groupby('User-ID').count()['Book-Rating'] > 200
rated_users = df2[df2].index
```

```python
filterd_users = df1[df1['User-ID'].isin(rated_users)]
```

```python
y = df1.groupby('Book-Title').count()['Book-Rating'] > 50
famous_books = filterd_users[filterd_users['Book-Title'].isin(y[y].index)]
```

```python
final_df = famous_books.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
final_df.fillna(0,inplace=True)
```

```python
final_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(final_df)
```

```python
def recommend(book_name):
    index = np.where(final_df.index == book_name)[0][0]
    similar_scores = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    new_indexes = [i[0] for i in similar_scores]
    
    suggestions = final_df.index[new_indexes]
    books = books_csv[books_csv['Book-Title'].isin(suggestions)]
    books = books.drop_duplicates('Book-Title')
    titles = books['Book-Title']
    authors = books['Book-Author']
    images = books['Image-URL-M']
    return zip(titles,authors,images)
```

```python
nl = sorted(list(enumerate(similarity_score[0])),key=lambda x:x[1],reverse=True)[1:6]
ids = []
for i in nl:
    ids.append(i[0])
ids
```
